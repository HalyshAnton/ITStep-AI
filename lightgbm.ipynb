Алгоритм дерева рішень належить до сімейства алгоритмів навчання з учителем. На відміну від інших алгоритмів навчання з учителем, дерево рішень може використовуватися для вирішення задач регресії та класифікації.

Мета дерева рішень полягає в тому, щоб створити навчальну модель, яка може прогнозувати клас (один або декілька) або значення, навчаючись простих правил прийняття рішень на основі навчальних даних.

Дерева рішень утворюють структуру, схожу на блок-схему, тому їх дуже легко інтерпретувати та розуміти. Це один з небагатьох алгоритмів машинного навчання, де дуже легко візуалізувати та проаналізувати внутрішню роботу алгоритму.

Так само, як і блок-схема, дерево рішень містить різні типи вузлів і гілок. Кожен вузол рішення представляє тест за ознакою, і на основі результату тесту він або утворює іншу гілку, або кінцевий вузол. Кожна гілка представляє правило прийняття рішення, а кінцевий вузол - кінцевий результат.

![дерево рішень](https://raw.githubusercontent.com/satishgunjal/images/master/Decision_Tree.png)


Типи дерев рішень
Дерева рішень бувають двох основних типів:

* **Дерева рішень класифікації**: У цьому типі дерев рішень цільова змінна є категоріальною. Наприклад, класифікація електронних листів як спаму або не спаму.
* **Дерева регресії**: У цьому типі дерев рішень цільова змінна є безперервною. Наприклад, прогнозування цін на будинки на основі їх характеристик.

# Внутрішня робота дерева рішень

Дерево рішень на кореневому вузлі обирає атрибут для розділення даних на дві основні категорії. Таким чином, наприкінці кореневого вузла ми маємо два правила рішення та два піддерева.

Дані знову будуть розділені на дві категорії в кожному піддереві. Цей процес триватиме, доки всі навчальні приклади не будуть згруповані разом.

Отже, наприкінці дерева рішень ми отримуємо кінцевий вузол. Він представляє клас або безперервне значення, яке ми намагаємося передбачити.

## Критерії розділення даних

Метою дерева рішень є розділення даних таким чином, щоб наприкінці ми отримали різні групи даних, які мають більшу схожість і меншу випадковість/нечистоту. Для досягнення цього кожен розділ у дереві рішень повинен зменшувати випадковість. Дерево рішень використовує критерії вибору "ентропія" або "коефіцієнт Джіні" для розділення даних.

Примітка: ми використовуватимемо бібліотеку scikit-learn для тестування класифікації та регресії. "Ентропія" або "коефіцієнт Джіні" є критеріями вибору для класифікатора, тоді як "mse" (середня квадратична похибка), "friedman_mse" та "mae" (середня абсолютна похибка) є критеріями вибору для регресора.

## Ентропія

Для того, щоб знайти найкращий атрибут, який зменшить випадковість після розділу, ми можемо порівняти випадковість до та після розділу для кожного атрибута. Зрештою, ми обираємо атрибут, який забезпечить найбільше зниження випадковості. Формально випадковість у даних називається "ентропією", а різниця між "ентропією" до та після розділу називається "коефіцієнтом інформаційного приросту". Оскільки в дереві рішень може бути кілька гілок, формулу коефіцієнта інформаційного приросту можна записати так:

![entropy_formula](https://raw.githubusercontent.com/satishgunjal/images/master/entropy_formula.png)

```
Коефіцієнт інформаційного приросту = Ентропія(Батьківський вузол) - (Середня ентропія(Дочірні вузли))
```

Таким чином, у випадку з "ентропією" дерево рішень розділить дані за допомогою атрибута, який забезпечує найбільший коефіцієнт інформаційного приросту.

## Коефіцієнт Джіні

У випадку нечистоти за Джіні, ми обираємо випадкову точку даних у нашому наборі даних. Потім випадково класифікуємо її відповідно до розподілу класів у наборі даних. Тому дуже важливо знати точність цієї випадкової класифікації. Нечистота за Джіні дає нам ймовірність неправильної класифікації. Ми визначимо якість розділу, зважуючи нечистоту кожної гілки кількістю елементів, які вона містить. Отримане значення називається "коефіцієнтом приросту Джіні" або "коефіцієнтом Джіні". Саме це використовується для вибору найкращого розділу в дереві рішень. Чим більший коефіцієнт приросту Джіні, тим краще розділення.

![gini_formula](https://raw.githubusercontent.com/satishgunjal/images/master/gini_formula.png)

Таким чином, у випадку з "коефіцієнтом Джіні" дерево рішень розділить дані за допомогою атрибута, який забезпечує найбільший коефіцієнт приросту Джіні.

## Що ж нам використовувати?

Нечистота за Джіні обчислюється швидше, оскільки не потребує обчислення логарифмічних функцій, хоча насправді жодна з метрик не призводить до більш точного дерева, ніж інша.

# Переваги дерева рішень

Дерева рішень є популярним алгоритмом машинного навчання завдяки низці переваг:

* **Легкість розуміння та інтерпретації.** Дерева можна візуалізувати, що дозволяє легко зрозуміти логіку прийняття ними рішень.
* **Мінімальна підготовка даних.** На відміну від інших методів, дерева рішень не потребують складної підготовки даних, такої як нормалізація, створення фіктивних змінних та видалення пустих значень.  Однак, зауважте, що цей модуль не підтримує роботу з відсутніми значеннями.
* **Здатність обробляти як числові, так і категоріальні дані.** Дерева рішень можуть працювати з різними типами даних, що робить їх універсальним інструментом.
* **Обробка задач з кількома вихідними значеннями.**  Дерева рішень можуть прогнозувати не тільки одне, але й декілька цільових значень одночасно.
* **Модель типу "біла скринька".** Оскільки логіка дерева рішень легко інтерпретується, ви можете зрозуміти, як воно приходить до своїх прогнозів.
* **Можливість перевірки моделі за допомогою статистичних тестів.** Ви можете використовувати статистичні тести для оцінки надійності моделі дерева рішень. Це дозволяє вам зрозуміти, наскільки можна довіряти її прогнозам.


# Недоліки дерева рішень

* **Перенавчання:** Алгоритми дерев рішень можуть створювати надто складні дерева, які погано узагальнюють дані. Це явище називається перенавчанням. Для уникнення цієї проблеми необхідні механізми, такі як обрізка гілок, встановлення мінімальної кількості зразків, необхідних у кінцевому вузлі, або встановлення максимальної глибини дерева.
* **Нестабільність:** Дерева рішень можуть бути нестабільними, оскільки незначні зміни в даних можуть призвести до створення абсолютно іншого дерева. Цю проблему можна пом'якшити за допомогою використання дерев рішень в ансамблі.
* **Зміщення в бік більшості класів:** Алгоритми дерев рішень створюють дерева із зміщенням, якщо деякі класи переважають інші. Тому перед навчанням за допомогою дерева рішень рекомендується балансувати набір даних.


# Що таке ансамблі?

![](https://prismic-io.s3.amazonaws.com/encord/1e02b8c6-cf23-46af-a6a9-f42678a00289_What+is+ensemble+learning%3F.png)

Ансамблі машинного навчання - це метод, який поєднує декілька моделей для покращення загальної точності та стійкості. Замість використання однієї моделі, ансамблі навчають декілька моделей на одному наборі даних і роблять прогнози, ґрунтуючись на комбінації виходів цих моделей. Це схоже на те, як люди збирають думки різних експертів, щоб прийняти краще рішення.

# Які є види ансамблів?

Існує декілька типів ансамблів, але найпоширеніші:

## Бегінг (Bootstrap Aggregation):

* Навчає декілька моделей на **різних підмножинах** даних, отриманих шляхом випадкового вибору ознак. Це допомагає зменшити залежність між моделями. Це схоже на ситуації, коли опитують декількох експертів, щоб прийняти остаточне рішення

## Бустинг

* Навчає декілька моделей **послідовно**. Кожна модель намагається виправити помилки попередніх моделей, роблячи невеликі корективи у прогнозах.

## Стекінг

* Поєднує прогнози декількох моделей, використовуючи метод регресії, щоб отримати остаточний прогноз.

# Бустинг та градієнтний бустинг

**Бустинг** - це тип ансамблю, де моделі навчаються послідовно. Кожна модель намагається виправити помилки попередніх моделей, роблячи невеликі корективи у прогнозах.

**Градієнтний бустинг** - це алгоритм бустингу, який використовує градієнтні методи для оптимізації прогнозів на кожному кроці. Це означає, що модель навчається на основі **псевдопомилок** або ж **градієнтів метрики** попередніх моделей, намагаючись мінімізувати загальну метрику.

![](https://cdn.analyticsvidhya.com/wp-content/uploads/2024/01/Gradient-Boosting-tree-2048x1152.jpg)

## Основні реалізації градієнтного бустингу

* **XGBoost:** Популярна бібліотека з відкритим кодом для градієнтного бустингу дерев.
* **CatBoost:** Бібліотека з відкритим кодом, яка використовує категоріальні ознаки більш ефективно, ніж XGBoost.
* **LightGBM:** Швидка та масштабована бібліотека з відкритим кодом для градієнтного бустингу дерев.

## Принцип роботи LightGBM

LightGBM використовує алгоритм градієнтного бустингу для навчання послідовності дерев. На кожному кроці:

1. Будується нове дерево, яке намагається мінімізувати **градієнт** помилки попередніх дерев.
2. Глибина та структура дерева регулюються за допомогою алгоритму "**листів за ростом**".
3. Використовується алгоритм "**Leaves-by-level**" для прискорення навчання.

![](https://media.licdn.com/dms/image/C5112AQG23dzGS3Q6qw/article-cover_image-shrink_600_2000/0/1523957272561?e=2147483647&v=beta&t=VpA1vj-4WJhF_qHrbItEZZM8TZ0QUsCpkqK7xjBnU2E)

## Переваги:

**Швидкість та масштабованість:**

* LightGBM використовує алгоритм "листів за ростом" та "Leaves-by-level", які значно прискорюють процес навчання.
* Завдяки своїй ефективності LightGBM може використовуватися для навчання на дуже великих наборах даних.

**Ефективна робота з категоріальними ознаками:**

* LightGBM може використовувати категоріальні ознаки без попереднього кодування, що робить його зручним для роботи з різними типами даних.
* Алгоритм "One-hot" кодування використовується лише для категорій, які використовуються при розділенні дерев.

**Висока точність:**

* LightGBM consistently achieves high accuracy on various machine learning benchmarks.
* Завдяки алгоритму градієнтного бустингу LightGBM може постійно вдосконалювати свої прогнози.

**Гнучкість:**

* LightGBM має багато параметрів, які можна налаштувати для отримання оптимальних результатів для конкретного завдання.
* Підтримує різні метрики та функції втрат.


## Недоліки:

**Складність налаштування:**

* Через велику кількість параметрів LightGBM може бути складним налаштувати для отримання найкращих результатів.
* Неправильне налаштування може призвести до перенавчання або недонавчання.

**Чутливість до шумних даних:**

* LightGBM може бути чутливим до шумних даних, тому перед навчанням важливо очистити дані.

**Можливість перенавчання:**

* Як і інші алгоритми машинного навчання, LightGBM може перенавчатися, якщо його не правильно налаштувати.
* Важливо використовувати методи регуляризації, такі як раннє зупинення або L1/L2 регуляризація, щоб запобігти перенавчанню.

**Інші:**

* LightGBM може бути менш виразним, ніж інші алгоритми, такі як XGBoost.
* Може вимагати більше пам'яті, ніж інші алгоритми.

LightGBM - це потужний алгоритм машинного навчання, який пропонує багато переваг, таких як швидкість, масштабованість, висока точність та гнучкість.

#Метрики

Метрики -- це числові характеристики для оцінки ефективності моделі:

* Середнє квадратичне відхилення (MSE)
* Корінь середньоквадратичного відхилення (RMSE).
* Середнє абсолютне відхилення (MAE).
* R² або коефіцієнт детермінації.
* Скоригований R²

### Середня абсолютне відхилення:

MAE — це модуль різниці між цільовим значенням і значенням, передбаченим моделлю.

MAE більш стійие до викидів і не штрафує за помилки так сильно, як mse.

MAE — це лінійна оцінка, яка означає, що всі індивідуальні відмінності зважені однаково.

Він не підходить для задач, де потрібно приділяти більше уваги викидам.

### Середнє квадратичне відхилення:

`MSE або середнє квадратичне відхилення` є однією із найбільш частовживаних метрик для завдань регресії.

Це просто середнє значення квадрата різниці між цільовим значенням і значенням, передбаченим регресійною моделлю.

Зводячи різницю в квадрат, він штрафує навіть невелику помилку, яка призводить до переоцінки того, наскільки погана модель.

Дану метрику часто використовують також тому, що вона є диференційованою(пізніше обговоримо).


<img src = "https://miro.medium.com/max/875/0*aTUPK_ILg7-n0znw.jpg">

### Корінь середньо квадратичного відхилення:

RMSE є найбільш широко використовуваною метрикою для задач регресії та є квадратним коренем із середньо квадратичного відхилення між цільовим значенням і значенням, передбаченим моделлю.

У деяких випадках це краще, тому що помилки спочатку зводяться в квадрат перед усередненням, що створює високий штраф за великі помилки.

Це означає, що RMSE є корисним, коли чутливість моделі до великих помилок небажана.

### R²

Коефіцієнт детермінації або R² — ще один показник, який використовується для оцінки ефективності регресійної моделі.

Метрика допомагає нам порівнювати наскільки краща наша модель за базову модель, яка завжди прогнозує середнє арифметичне, або медіану.

R² – це безмасштабна оцінка, яка означає, що значення R² завжди буде меншим або дорівнює 1, не має значення, чи є значення занадто великими чи занадто малими.

Значення в 1 означає, що модель хороша, 0 - що така сама як базова

<img src = "https://miro.medium.com/max/783/0*_Bk3m941thWlveS3.png">

### Скоригований R²:

Скоригований R² має те саме значення, що й R², але є його вдосконаленням.

R² страждає від проблеми, пов’язаної з підвищенням значенням при збільшені кількості незалежних ознак, навіть якщо модель не покращується, що може ввести дослідника в оману.

Скоригований R² завжди нижчий за R², оскільки він коригується з урахуванням кількості ознак і показує покращення, лише якщо є реальне покращення.

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import model_selection, metrics
import lightgbm as lgb
````

# Побудова моделі

## Поділ даних

````python
df.columns
````

````python
y = df['Profit']
X = df.drop(columns='Profit')
````

````python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    train_size=0.8,
                                                    random_state=42)
````

## Створення Pipeline

````python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OrdinalEncoder
````

````python
num_columns = X.select_dtypes(include="number").columns
cat_columns = X.select_dtypes(include="object").columns

# числові ознаки
num_transformer = Pipeline(
    steps=[("imputer", SimpleImputer(strategy="median"))]
    )

# категоріальні ознаки
cat_transformer = Pipeline(
    steps=[("imputer", SimpleImputer(strategy="most_frequent")),
           ("encoder", OrdinalEncoder(handle_unknown='use_encoded_value',
                                      unknown_value=-1))
           ])

# об'єднання двох pipeline
preprocessor = ColumnTransformer(
    transformers=[
        ("num", num_transformer, num_columns),
        ("cat", cat_transformer, cat_columns),
    ]
)

preprocessor.set_output(transform='pandas')
preprocessor
````

# Попередня обробка даних

Оскільки ми користуємось новою бібліотекою, то нам потрібно поередньо натренувати `preprocessor` та обробити дані

````python
preprocessor.fit(X_train)

X_train_prep = preprocessor.transform(X_train)
X_test_prep = preprocessor.transform(X_test)

X_train_prep.head(5)
````

## Основна модель

````python
import lightgbm as lgbm

model = lgbm.LGBMRegressor(objective='mse',
                           n_estimators=300,
                           learning_rate=0.1
                           )
````

### Опис параметрів:

* **objective:** Тип завдання, для якого використовується модель. Може бути "regression" (регресія) або "binary" (бінарна класифікація) чи якась інша **функція втрат**
* **boosting_type:** Тип бустингу, який використовується для побудови моделі. LightGBM підтримує кілька типів бустингу:
 * `gbdt` (за замовчуванням): Градієнтний бустинг з деревами рішень (GBDT) - базова реалізація градієнтного бустингу.
 * `dart`: Дерева з диференційованими агрегатами (Differentiable Automatic Relevance Determination, DART). Цей тип бустингу використовує більш гладкі функції розподілу для дерев, що дозволяє використовувати оптимізатори, засновані на градієнтах.
 * `rf`: Випадковий ліс (Random Forest).
* **n_estimators:** Кількість дерев в ансамблі. Це один з найважливіших параметрів, який впливає на точність та складність моделі.
* **learning_rate:** Швидкість навчання, яка визначає, як швидко модель оновлює свої ваги на кожному кроці. Більш висока швидкість навчання може призвести до кращої точності, але також може збільшити ризик перенавчання.
* **num_leaves:** Максимальна кількість листків у дереві. Цей параметр контролює складність кожного дерева. Більша кількість листків може призвести до більш складних дерев, але також може збільшити ризик перенавчання.
* **feature_fraction:** Частка ознак, що використовуються при розділенні в дереві. Цей параметр допомагає запобігти перенавчанню та покращити узагальнення моделі.
* **bagging_fraction:** Частка даних, що використовується для навчання кожного дерева. Цей параметр використовується для створення різних підмножин даних для навчання кожного дерева, що покращує стійкість моделі.
* **bagging_freq:** Частота, з якою використовується `bagging_fraction`. Більш високе значення `bagging_freq` призводить до більшої різноманітності дерев.
* **min_child_samples:** Мінімальна кількість зразків у кінцевому вузлі. Цей параметр використовується для запобігання створенню надто маленьких дерев.
* **min_child_weight:** Мінімальна вага кінцевого вузла. Цей параметр використовується для запобігання створенню дерев з надто легкими кінцевими вузлами.
* **lambda_l1:** Коефіцієнт L1 регуляризації.
* **lambda_l2:** Коефіцієнт L2 регуляризації.

Додаткові параметри:

* **n_jobs:** (int, default=-1) - Кількість потоків.
* **silent:** (bool, default=False) - Чи виводити повідомлення про
хід навчання.
* **seed:** (int, default=None) -
Випадкове насіння.
* **feature_selector:** (str or callable, default=None) - Селектор ознак.
* **cat_features:** (list, default=None) - Катеґорійні ознаки.
* **group_features:** (list, default=None) - Групові ознаки.
* **application:** (str, default="regression") - Тип завдання.


````python
model.fit(X_train_prep, y_train,
          eval_set=[(X_train_prep, y_train),
                    (X_test_prep, y_test)],
          eval_names=['train', 'test'],
          eval_metric=['mae'])
````

### Опис методу `fit()`

Метод `fit()` в LightGBM використовується для навчання моделі на основі даних та міток. Він приймає декілька параметрів, які впливають на процес навчання:

**Основні параметри:**

* **data:** Набір даних, який використовується для навчання моделі. Це може бути матриця NumPy або DataFrame pandas.
* **eval_set:** Набір даних для оцінки якості моделі під час навчання. Це може бути матриця NumPy або DataFrame pandas.
* **eval_metric:** Метрика, яка використовується для оцінки якості моделі, але на якій вона не буде начатись. Доступні метрики: "rmse", "mae", "auc", "precision", "recall", "f1", "accuracy", "multi_class_logloss", "multi_class_accuracy".

**Додаткові параметри:**

* **verbose:** Рівень деталізації виводу під час навчання. Можливі значення: 0 (немає виводу), 1 (короткий вивід), 2 (детальний вивід).
* **early_stopping_rounds:** Кількість ітерацій без покращення метрики на `eval_set`, після яких навчання буде зупинено.

### Проблема з R2

Custom Eval Function використовується для визначення власної метрики оцінки якості моделі в LightGBM. Ця функція може бути корисною, коли стандартні метрики, такі як MSE або RMSE, не підходять для вашого завдання.

### Структура функції

Така функція повинна приймати два аргументи:

* **y_true:** справжні значення
* **y_pred** результат спрогнозований моделлю

Функція повинна повертати кортеж з трьох елементів:

* **Назва метрики:** Строка, яка описує метрику.
* **Значення метрики:** Числове значення метрики.
* **is_higher_better:** Логічне значення, яке вказує, чи краще більше значення метрики (True) або нижче (False).

````python
def lgbm_r2(y_true, y_pred):
    r2 = metrics.r2_score(y_true, y_pred)
    return "r2", r2, True
````

# Метрики

````python
from sklearn import metrics

def get_metrics(y_true, y_pred, name="model"):
  df = pd.DataFrame()

  df.loc["MAE", name] = metrics.mean_absolute_error(y_true, y_pred)
  df.loc["MSE", name] = metrics.mean_squared_error(y_true, y_pred)
  df.loc["RMSE", name] = metrics.mean_squared_error(y_true, y_pred) ** 0.5
  df.loc["R2", name] = metrics.r2_score(y_true, y_pred)

  return df.round(2)
````

````python
all_metrics = pd.DataFrame()

y_pred = model.predict(X_train_prep)
all_metrics['train with mse'] = get_metrics(y_train, y_pred)

y_pred = model.predict(X_test_prep)
all_metrics['test with mse'] = get_metrics(y_test, y_pred)

all_metrics
````
