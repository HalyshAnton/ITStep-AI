# –ó–º—ñ–Ω–∞ transformer –ø—ñ—Å–ª—è –ø–æ–¥—ñ–ª—É

* –°—Ç–≤–æ—Ä—é—î–º–æ TransformDataset, —è–∫–∏–π –∑–∞—Å—Ç–æ—Å–æ–≤—É–≤–∞—Ç–∏–º–µ –ø–µ—Ä–µ—Ç–≤–æ—Ä–µ–Ω–Ω—è –¥–æ –∑–æ–±—Ä–∞–∂–µ–Ω—å.
* –ö–ª–∞—Å `TransformDataset` —É—Å–ø–∞–¥–∫–æ–≤—É—î—Ç—å—Å—è –≤—ñ–¥ `torch.utils.data.Dataset` —ñ –≤–∏–∑–Ω–∞—á–∞—î –º–µ—Ç–æ–¥–∏ –¥–æ—Å—Ç—É–ø—É –¥–æ –¥–∞–Ω–∏—Ö.
* `__len__` –ø–æ–≤–µ—Ä—Ç–∞—î —Ä–æ–∑–º—ñ—Ä –Ω–∞–±–æ—Ä—É –¥–∞–Ω–∏—Ö.
* `__getitem__` –ø–æ–≤–µ—Ä—Ç–∞—î –≤–∏–±—ñ—Ä–∫—É –¥–∞–Ω–∏—Ö (–æ–∑–Ω–∞–∫–∏ —Ç–∞ —Ü—ñ–ª—å) –∑–∞ –ø–µ–≤–Ω–∏–º —ñ–Ω–¥–µ–∫—Å–æ–º.

````python
from torch.utils.data import Dataset

class TransformDataset(Dataset):
    def __init__(self, dataset, transformer):
        super().__init__()
        self.dataset = dataset
        self.transformer = transformer

    def __len__(self):
        return len(self.dataset)
    
    def __getitem__(self, idx):
        img, y = self.dataset[idx]

        new_img = self.transformer(img)
        
        return new_img, y
````

# –ü—Ä–∏–∫–ª–∞–¥ —Å—Ç–≤–æ—Ä–µ–Ω–Ω—è –≤–¥–∞—Å–Ω–æ–≥–æ Dataset

````python
import os
from PIL import Image


class ImageDataset(torch.utils.data.Dataset):
    """
    Custom Dataset for loading and preprocessing images.
    """
    def __init__(self, root_dir, transform=None, task='Train'):
        """
        Args:
          root_dir (str): Path to the directory containing images.
          transform (torchvision.transforms, optional): Transformations to apply to images. Defaults to None.
        """
        self.root_dir = root_dir
        self.image_paths = self.get_pathes(root_dir, task)

        self.classes = os.listdir(root_dir)
        self.class_to_idx = {label: i for i, label in enumerate(self.classes)}
        self.transform = transform

    def get_pathes(self, root, task):
        image_paths = []
        labels = os.listdir(root)

        for label in labels:
            images = os.listdir(os.path.join(root, label, task))    
            image_paths.extend([os.path.join(root, label, task, image) for image in images])
        return image_paths

    def __len__(self):
        """
        Returns the number of images in the dataset.
        """
        return len(self.image_paths)

    def __getitem__(self, idx):
        """
        Loads and preprocesses an image at a given index.

        Args:
          idx (int): Index of the image to return.

        Returns:
          tuple: A tuple containing the preprocessed image and its label (if available).
        """
        image_path = self.image_paths[idx]
        image = Image.open(image_path).convert('RGB')  # Assuming RGB images
        if self.transform:
            image = self.transform(image)

        # Add logic to load labels if available (modify based on your data structure)
        label_name = os.path.normpath(image_path).split(os.path.sep)[-3]

        return image, self.class_to_idx[label_name]
````

## ReLU

* **–§—É–Ω–∫—Ü—ñ—è:** f(x) = max(0, x)
* **–ü–µ—Ä–µ–≤–∞–≥–∏:** –ü—Ä–æ—Å—Ç–∞, –æ–±—á–∏—Å–ª—é–≤–∞–ª—å–Ω–æ –µ—Ñ–µ–∫—Ç–∏–≤–Ω–∞, –µ—Ñ–µ–∫—Ç–∏–≤–Ω–∞ –¥–ª—è —Ä—ñ–∑–Ω–∏—Ö –∑–∞–¥–∞—á.
* **–ù–µ–¥–æ–ª—ñ–∫–∏:** –ú–æ–∂–µ —Å—Ç—Ä–∞–∂–¥–∞—Ç–∏ –≤—ñ–¥ –ø—Ä–æ–±–ª–µ–º–∏ "–≤–º–∏—Ä–∞–Ω–Ω—è ReLU", –∫–æ–ª–∏ –Ω–µ–π—Ä–æ–Ω–∏ —Å—Ç–∞—é—Ç—å –ø–æ—Å—Ç—ñ–π–Ω–æ –Ω–µ–∞–∫—Ç–∏–≤–Ω–∏–º–∏ –¥–ª—è –≤—ñ–¥'—î–º–Ω–∏—Ö –≤—Ö—ñ–¥–Ω–∏—Ö –¥–∞–Ω–∏—Ö.

* **–ü–µ—Ä–µ–≤–∞–≥–∏:** –í–∏—Ä—ñ—à—É—î –ø—Ä–æ–±–ª–µ–º—É "–≤–º–∏—Ä–∞—é—á–æ—ó ReLU", –¥–æ–∑–≤–æ–ª—è—é—á–∏ –Ω–µ–≤–µ–ª–∏–∫–∏–π –¥–æ–¥–∞—Ç–Ω–∏–π –≥—Ä–∞–¥—ñ—î–Ω—Ç –¥–ª—è –≤—ñ–¥'—î–º–Ω–∏—Ö –≤—Ö–æ–¥—ñ–≤.
* **–ù–µ–¥–æ–ª—ñ–∫–∏:** –î–µ—â–æ —Å–∫–ª–∞–¥–Ω—ñ—à–∏–π –∑–∞ ReLU.

## ELU

* **–§—É–Ω–∫—Ü—ñ—è:** f(x) =
  \begin{cases}
    \beta \cdot (\exp(x) - 1) & \text{if } x < 0 \\
    x & \text{—ñ–Ω–∞–∫—à–µ}
  \end{cases}
    (–¥–µ Œ≤ - –≥—ñ–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä)
* **–ü–µ—Ä–µ–≤–∞–≥–∏:** –ü–ª–∞–≤–Ω–∏–π –ø–µ—Ä–µ—Ö—ñ–¥ –ø—Ä–∏ x = 0, —É–Ω–∏–∫–∞—î "–≤–º–∏—Ä–∞–Ω–Ω—è ReLU", –º–∞—î —Ç–µ–Ω–¥–µ–Ω—Ü—ñ—é –¥–æ —à–≤–∏–¥—à–æ—ó –∑–±—ñ–∂–Ω–æ—Å—Ç—ñ.
* **–ù–µ–¥–æ–ª—ñ–∫–∏:** –î–µ—â–æ –¥–æ—Ä–æ–∂—á–∏–π –≤ –æ–±—á–∏—Å–ª—é–≤–∞–ª—å–Ω–æ–º—É –ø–ª–∞–Ω—ñ, –Ω—ñ–∂ ReLU.

## GELU

* **–§—É–Ω–∫—Ü—ñ—è:** $f(x) = 0.5x \left(1 + \text{erf} \left( \frac{x}{\sqrt{2}} \right)\right)$

 (–¥–µ erf - —Ñ—É–Ω–∫—Ü—ñ—è –ø–æ—Ö–∏–±–∫–∏)
* **–ü–µ—Ä–µ–≤–∞–≥–∏:** –ü–ª–∞–≤–Ω–æ –Ω–∞–±–ª–∏–∂–∞—î —Ñ—É–Ω–∫—Ü—ñ—é —Ç–æ—Ç–æ–∂–Ω–æ—Å—Ç—ñ –¥–ª—è –¥–æ–¥–∞—Ç–Ω–∏—Ö –∑–Ω–∞—á–µ–Ω—å, —É–Ω–∏–∫–∞—î "–≤–º–∏—Ä–∞–Ω–Ω—è ReLU".
* **–ù–µ–¥–æ–ª—ñ–∫–∏:** –û–±—á–∏—Å–ª—é–≤–∞–ª—å–Ω–æ –¥–æ—Ä–æ–∂—á–∏–π –∑–∞ ReLU —á–µ—Ä–µ–∑ —Ñ—É–Ω–∫—Ü—ñ—é erf.

# –°—Ç–≤–æ—Ä–µ–Ω–Ω—è –º–æ–¥–µ–ª—ñ

## –Ø–∫ –≤–∏–∑–Ω–∞—á–∏—Ç–∏ –∫—ñ–ª—å–∫—ñ—Å—Ç—å –Ω–µ–π—Ä–æ–Ω—ñ–≤ —É –ø–µ—Ä—à–æ–º—É —à–∞—Ä—ñ –¥–ª—è –∫–ª–∞—Å–∏—Ñ—ñ–∫–∞—Ü—ñ—ó –∑–æ–±—Ä–∞–∂–µ–Ω—å?

–£ –ø–µ—Ä—à–æ–º—É (–≤—Ö—ñ–¥–Ω–æ–º—É) —à–∞—Ä—ñ –Ω–µ–π—Ä–æ–Ω–Ω–æ—ó –º–µ—Ä–µ–∂—ñ –∫—ñ–ª—å–∫—ñ—Å—Ç—å –Ω–µ–π—Ä–æ–Ω—ñ–≤ **–∑–∞–ª–µ–∂–∏—Ç—å –≤—ñ–¥ —Ä–æ–∑–º—ñ—Ä—É –∑–æ–±—Ä–∞–∂–µ–Ω–Ω—è**.

### –ü—Ä–∞–≤–∏–ª–æ:
> **–ö—ñ–ª—å–∫—ñ—Å—Ç—å –Ω–µ–π—Ä–æ–Ω—ñ–≤ = –∫—ñ–ª—å–∫—ñ—Å—Ç—å –ø—ñ–∫—Å–µ–ª—ñ–≤ —É –∑–æ–±—Ä–∞–∂–µ–Ω–Ω—ñ**

–¶–µ —Ç–æ–º—É, —â–æ –∫–æ–∂–µ–Ω –Ω–µ–π—Ä–æ–Ω —É –≤—Ö—ñ–¥–Ω–æ–º—É —à–∞—Ä—ñ –æ—Ç—Ä–∏–º—É—î –∑–Ω–∞—á–µ–Ω–Ω—è **–æ–¥–Ω–æ–≥–æ –ø—ñ–∫—Å–µ–ª—è** (–∞–±–æ –∫–∞–Ω–∞–ª—É –ø—ñ–∫—Å–µ–ª—è) –∑–æ–±—Ä–∞–∂–µ–Ω–Ω—è.

### –ü—Ä–∏–∫–ª–∞–¥–∏:

- üñºÔ∏è –ó–æ–±—Ä–∞–∂–µ–Ω–Ω—è 28√ó28 –ø—ñ–∫—Å–µ–ª—ñ–≤, —á–æ—Ä–Ω–æ-–±—ñ–ª–µ (1 –∫–∞–Ω–∞–ª):  
  `28 √ó 28 √ó 1 = 784 –Ω–µ–π—Ä–æ–Ω–∏`

- üñºÔ∏è –ó–æ–±—Ä–∞–∂–µ–Ω–Ω—è 64√ó64, –∫–æ–ª—å–æ—Ä–æ–≤–µ (3 –∫–∞–Ω–∞–ª–∏ ‚Äî RGB):  
  `64 √ó 64 √ó 3 = 12 288 –Ω–µ–π—Ä–æ–Ω—ñ–≤`


````python
img, y = train_dataset[0]
img.shape
````

````python
from torch import nn


model = nn.Sequential(
    nn.Flatten(),
    nn.Linear(64*64*3, 200),
    nn.ReLU(),
    nn.Linear(200, 64),
    nn.LeakyReLU(),
    nn.Linear(64, 32),
    nn.ReLU(),
    nn.Linear(32, 10)
    
)

# –ø—ñ–¥–∫–ª—é—á–µ–Ω–Ω—è –¥–æ –ø—Ä–æ—Ü–µ—Å–æ—Ä–∞
model = model.to(device)
model
````

# –í–∏–∑–Ω–∞—á–µ–Ω–Ω—è —Ñ—É–Ω–∫—Ü—ñ—ó –≤—Ç—Ä–∞—Ç —Ç–∞ –æ–ø—Ç–∏–º—ñ–∑–∞—Ç–æ—Ä–∞

````python
# –§—É–Ω–∫—Ü—ñ—è –≤—Ç—Ä–∞—Ç –¥–ª—è –∫–ª–∞—Å–∏—Ñ—ñ–∫–∞—Ü—ñ—ó
loss_fn = nn.CrossEntropyLoss()

# –û–ø—Ç–∏–º—ñ–∑–∞—Ç–æ—Ä (Adam) –¥–ª—è –æ–Ω–æ–≤–ª–µ–Ω–Ω—è –≤–∞–≥ –º–æ–¥–µ–ª—ñ
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)
````

````python
epochs = 10 # –∫—ñ–ª—å–∫—ñ—Å—Ç—å —ñ—Ç–µ—Ä–∞—Ü—ñ–π –¥–ª—è —Ç—Ä–µ–Ω—É–≤–∞–Ω–Ω—è

for i in range(epochs):
    # –Ω–∞–≤—á–∞–Ω–Ω—ñ –Ω–∞ —É—Å—ñ—Ö –¥–∞–Ω–∏—Ö
    losses_ = []
    for X_batch, y_batch in loader:
        # –ø—ñ–¥–∫–ª—é—á–µ–Ω–Ω—è –¥–æ –ø—Ä–æ—Ü–µ—Å–æ—Ä–∞
        X_batch = X_batch.to(device)
        y_batch = y_batch.to(device)

        # –Ω–∞–≤—á–∞–Ω–Ω—è –Ω–∞ –æ–¥–Ω–æ–º—É –ø–∞–∫–µ—Ç—ñ

        y_pred = model(X_batch)
        loss = loss_fn(y_pred, y_batch)  # –ø–æ–º–∏–ª–∫–∞ —è–∫—É —Ç—Ä–µ–±–∞ –∑—Ä–æ–±–∏—Ç–∏ —è–∫–æ–º–æ–≥–∞ –º–µ–Ω—à–æ—é
        losses_.append(loss.item())

        # —Ä–∞—Ö—É—î–º–æ –≥—Ä–∞–¥—ñ—î–Ω—Ç–∏ —ñ –∑–∞—Å—Ç–æ—Å–æ–≤—É—î–º–æ –æ–ø—Ç–∏–º—ñ–∑–∞—Ç–æ—Ä
        loss.backward()
        optimizer.step()
        optimizer.zero_grad()

    print(f"{i} -- {np.mean(losses_)}")
````

–£—è–≤—ñ—Ç—å —Å–æ–±—ñ, —è–∫ –∑–±—ñ–ª—å—à—É–≤–∞–ª—å–Ω–µ —Å–∫–ª–æ –¥–µ—Ç–µ–∫—Ç–∏–≤–∞ —Ä–µ—Ç–µ–ª—å–Ω–æ —Å–∫–∞–Ω—É—î –º—ñ—Å—Ü–µ –∑–ª–æ—á–∏–Ω—É –≤ –ø–æ—à—É–∫–∞—Ö –¥–æ–∫–∞–∑—ñ–≤. –£ –∫–æ–º–ø'—é—Ç–µ—Ä–Ω–æ–º—É –∑–æ—Ä—ñ –∑–≥–æ—Ä—Ç–∫–∞ –¥—ñ—î –∞–Ω–∞–ª–æ–≥—ñ—á–Ω–æ. –¶–µ –º–∞—Ç–µ–º–∞—Ç–∏—á–Ω–∞ –æ–ø–µ—Ä–∞—Ü—ñ—è, —è–∫–∞ –≤–∫–ª—é—á–∞—î —Ñ—ñ–ª—å—Ç—Ä (—è–¥—Ä–æ), –ø–æ–¥—ñ–±–Ω–µ –¥–æ –º—ñ–Ω—ñ–∞—Ç—é—Ä–Ω–æ–≥–æ –∑–æ–±—Ä–∞–∂–µ–Ω–Ω—è, —â–æ –∫–æ–≤–∑–∞—î –ø–æ –≤—Ö—ñ–¥–Ω–æ–º—É –∑–æ–±—Ä–∞–∂–µ–Ω–Ω—é. –£ –∫–æ–∂–Ω–æ–º—É –º—ñ—Å—Ü—ñ, –¥–µ –≤—ñ–Ω –ø–æ—Ç—Ä–∞–ø–ª—è—î, —Ñ—ñ–ª—å—Ç—Ä –ø–æ–µ–ª–µ–º–µ–Ω—Ç–Ω–æ –º–Ω–æ–∂–∏—Ç—å –≤—ñ–¥–ø–æ–≤—ñ–¥–Ω—ñ –ø—ñ–∫—Å–µ–ª—ñ –Ω–∞ –∑–æ–±—Ä–∞–∂–µ–Ω–Ω—ñ —Ç–∞ –ø—ñ–¥—Å—É–º–æ–≤—É—î –¥–æ–±—É—Ç–∫–∏. –¶–µ –≥–µ–Ω–µ—Ä—É—î –æ–¥–Ω–µ –∑–Ω–∞—á–µ–Ω–Ω—è, —â–æ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª—è—î —Ä–µ–∞–∫—Ü—ñ—é —Ñ—ñ–ª—å—Ç—Ä–∞ –Ω–∞ –ø–µ–≤–Ω—É –æ–±–ª–∞—Å—Ç—å –∑–æ–±—Ä–∞–∂–µ–Ω–Ω—è. –°–∏—Å—Ç–µ–º–∞—Ç–∏—á–Ω–æ –ø–µ—Ä–µ–º—ñ—â—É—é—á–∏ —Ñ—ñ–ª—å—Ç—Ä –ø–æ –≤—Å—å–æ–º—É –∑–æ–±—Ä–∞–∂–µ–Ω–Ω—é, –º–∏ –æ—Ç—Ä–∏–º—É—î–º–æ –Ω–æ–≤–∏–π –¥–≤–æ–≤–∏–º—ñ—Ä–Ω–∏–π –º–∞—Å–∏–≤, —è–∫–∏–π –Ω–∞–∑–∏–≤–∞—î—Ç—å—Å—è –∫–∞—Ä—Ç–æ—é —Ñ—É–Ω–∫—Ü—ñ–π.

# Softmax: –ü–µ—Ä–µ—Ç–≤–æ—Ä–µ–Ω–Ω—è —Å–∏—Ä–∏—Ö –æ—Ü—ñ–Ω–æ–∫ –Ω–∞ –π–º–æ–≤—ñ—Ä–Ω–æ—Å—Ç—ñ

–£ —Å—Ñ–µ—Ä—ñ –º–∞—à–∏–Ω–Ω–æ–≥–æ –Ω–∞–≤—á–∞–Ω–Ω—è, –æ—Å–æ–±–ª–∏–≤–æ –≤ –∑–∞–¥–∞—á–∞—Ö –±–∞–≥–∞—Ç–æ–∫–ª–∞—Å–æ–≤–æ—ó –∫–ª–∞—Å–∏—Ñ—ñ–∫–∞—Ü—ñ—ó, —Ñ—É–Ω–∫—Ü—ñ—è softmax –≤—ñ–¥—ñ–≥—Ä–∞—î –≤–∏—Ä—ñ—à–∞–ª—å–Ω—É —Ä–æ–ª—å. –í–æ–Ω–∞ –±–µ—Ä–µ –≤–µ–∫—Ç–æ—Ä –¥—ñ–π—Å–Ω–∏—Ö —á–∏—Å–µ–ª (—á–∞—Å—Ç–æ —Ü–µ –Ω–µ–æ–±—Ä–æ–±–ª–µ–Ω—ñ –æ—Ü—ñ–Ω–∫–∏ –Ω–µ–π—Ä–æ–Ω–Ω–æ—ó –º–µ—Ä–µ–∂—ñ) —ñ –ø–µ—Ä–µ—Ç–≤–æ—Ä—é—î –π–æ–≥–æ —É –≤–µ–∫—Ç–æ—Ä –π–º–æ–≤—ñ—Ä–Ω–æ—Å—Ç–µ–π, —è–∫—ñ –≤ —Å—É–º—ñ –¥–æ—Ä—ñ–≤–Ω—é—é—Ç—å 1. –¶—ñ –π–º–æ–≤—ñ—Ä–Ω–æ—Å—Ç—ñ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª—è—é—Ç—å –π–º–æ–≤—ñ—Ä–Ω—ñ—Å—Ç—å —Ç–æ–≥–æ, —â–æ –≤—Ö—ñ–¥–Ω—ñ –¥–∞–Ω—ñ –Ω–∞–ª–µ–∂–∞—Ç—å –¥–æ –∫–æ–∂–Ω–æ–≥–æ –º–æ–∂–ª–∏–≤–æ–≥–æ –∫–ª–∞—Å—É.

**–í–ª–∞—Å—Ç–∏–≤–æ—Å—Ç—ñ –≤–∏—Ö–æ–¥—ñ–≤ Softmax:**

* **–ó–Ω–∞—á–µ–Ω–Ω—è –º—ñ–∂ 0 —Ç–∞ 1:** –ö–æ–∂–µ–Ω –≤–∏—Ö—ñ–¥ –ø—ñ—Å–ª—è –ø–µ—Ä–µ—Ç–≤–æ—Ä–µ–Ω–Ω—è softmax –ª–µ–∂–∏—Ç—å –º—ñ–∂ 0 —Ç–∞ 1, –ø—Ä–µ–¥—Å—Ç–∞–≤–ª—è—é—á–∏ –π–º–æ–≤—ñ—Ä–Ω—ñ—Å—Ç—å.
* **–ù–∞–π–≤–∏—â–∞ –π–º–æ–≤—ñ—Ä–Ω—ñ—Å—Ç—å –≤–∫–∞–∑—É—î –Ω–∞ –ø–µ—Ä–µ–¥–±–∞—á—É–≤–∞–Ω–∏–π –∫–ª–∞—Å:** –ï–ª–µ–º–µ–Ω—Ç –∑ –Ω–∞–π–≤–∏—â–æ—é –π–º–æ–≤—ñ—Ä–Ω—ñ—Å—Ç—é —É –≤–∏—Ö—ñ–¥–Ω–æ–º—É –≤–µ–∫—Ç–æ—Ä—ñ –≤–≤–∞–∂–∞—î—Ç—å—Å—è –Ω–∞–π–±—ñ–ª—å—à –π–º–æ–≤—ñ—Ä–Ω–∏–º –∫–ª–∞—Å–æ–º –¥–ª—è –≤—Ö—ñ–¥–Ω–∏—Ö –¥–∞–Ω–∏—Ö.

**–ó–∞—Å—Ç–æ—Å—É–≤–∞–Ω–Ω—è Softmax:**

* **–ë–∞–≥–∞—Ç–æ–∫–ª–∞—Å–æ–≤–∞ –∫–ª–∞—Å–∏—Ñ—ñ–∫–∞—Ü—ñ—è:** Softmax —î –≤–∞–∂–ª–∏–≤–∏–º —É —Ç–∞–∫–∏—Ö –∑–∞–¥–∞—á–∞—Ö, —è–∫ —Ä–æ–∑–ø—ñ–∑–Ω–∞–≤–∞–Ω–Ω—è –∑–æ–±—Ä–∞–∂–µ–Ω—å (–∫–ª–∞—Å–∏—Ñ—ñ–∫–∞—Ü—ñ—è –æ–±'—î–∫—Ç—ñ–≤ –Ω–∞ –∑–æ–±—Ä–∞–∂–µ–Ω–Ω—è—Ö), –∞–Ω–∞–ª—ñ–∑ –Ω–∞—Å—Ç—Ä–æ—ó–≤ (–∫–ª–∞—Å–∏—Ñ—ñ–∫–∞—Ü—ñ—è —Ç–µ–∫—Å—Ç—É —è–∫ –ø–æ–∑–∏—Ç–∏–≤–Ω–æ–≥–æ, –Ω–µ–≥–∞—Ç–∏–≤–Ω–æ–≥–æ –∞–±–æ –Ω–µ–π—Ç—Ä–∞–ª—å–Ω–æ–≥–æ) —Ç–∞ –≤–∏—è–≤–ª–µ–Ω–Ω—è —Å–ø–∞–º—É (–∫–ª–∞—Å–∏—Ñ—ñ–∫–∞—Ü—ñ—è –µ–ª–µ–∫—Ç—Ä–æ–Ω–Ω–∏—Ö –ª–∏—Å—Ç—ñ–≤ —è–∫ —Å–ø–∞–º—É –∞–±–æ –Ω–µ —Å–ø–∞–º—É).
* **–í–∏—Ö—ñ–¥–Ω–∏–π —à–∞—Ä –Ω–µ–π—Ä–æ–Ω–Ω–∏—Ö –º–µ—Ä–µ–∂:** –£ –∑–∞–¥–∞—á–∞—Ö –±–∞–≥–∞—Ç–æ–∫–ª–∞—Å–æ–≤–æ—ó –∫–ª–∞—Å–∏—Ñ—ñ–∫–∞—Ü—ñ—ó —Ñ—É–Ω–∫—Ü—ñ—è softmax –∑–∞–∑–≤–∏—á–∞–π –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—î—Ç—å—Å—è —è–∫ —Ñ—É–Ω–∫—Ü—ñ—è –∞–∫—Ç–∏–≤–∞—Ü—ñ—ó –≤ –æ—Å—Ç–∞–Ω–Ω—å–æ–º—É —à–∞—Ä—ñ –Ω–µ–π—Ä–æ–Ω–Ω–æ—ó –º–µ—Ä–µ–∂—ñ. –ú–µ—Ä–µ–∂–∞ –≤–∏–≤–æ–¥–∏—Ç—å —Å–∏—Ä—ñ –æ—Ü—ñ–Ω–∫–∏, –∞ softmax –ø–µ—Ä–µ—Ç–≤–æ—Ä—é—î —ó—Ö –Ω–∞ –π–º–æ–≤—ñ—Ä–Ω–æ—Å—Ç—ñ, —è–∫—ñ –º–æ–∂–Ω–∞ —ñ–Ω—Ç–µ—Ä–ø—Ä–µ—Ç—É–≤–∞—Ç–∏.

````python
import numpy as np
from torch import nn
import torch.nn.functional as F


class MyModel(nn.Module):
    def __init__(self):
        super().__init__()
        
        self.conv1 = nn.Conv2d(in_channels=3, out_channels=8, kernel_size=3, padding='same')
        self.conv2 = nn.Conv2d(in_channels=8, out_channels=8, kernel_size=3, padding='same')
        self.conv3 = nn.Conv2d(in_channels=8, out_channels=16, kernel_size=3, padding='same')
        self.conv4 = nn.Conv2d(in_channels=16, out_channels=16, kernel_size=3, padding='same')
        self.conv5 = nn.Conv2d(in_channels=16, out_channels=16, kernel_size=3, padding='same')

        self.pool = nn.MaxPool2d(kernel_size=2, stride=2)
        self.flat = nn.Flatten()

        self.linear1 = nn.Linear(1024, 128)
        self.linear2 = nn.Linear(128, 10)

    def forward(self, x):
        # x -- (batch, 3, 64, 64)
        out = self.conv1(x)  # (batch, 8, 64, 64)
        out = F.relu(out)

        out = self.conv2(out) # (batch, 8, 64, 64)
        out = F.relu(out)

        out = self.pool(out) # (batch, 8, 32, 32)

        out = self.conv3(out)  # (batch, 16, 32, 32)
        out = F.relu(out)

        out = self.conv4(out) # (batch, 16, 32, 32)
        out = F.relu(out)

        out = self.pool(out) # (batch, 16, 16, 16)

        out = self.conv5(out) # (batch, 16, 16, 16)
        out = F.relu(out)

        out = self.pool(out) # (batch, 16, 8, 8)
        out = self.flat(out) # (batch, 1024)

        out = self.linear1(out)
        out = F.relu(out)

        out = self.linear2(out)

        return out


model = MyModel()
````
